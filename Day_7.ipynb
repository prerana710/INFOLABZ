<h1> <center> Multiple Linear Regreassion (MLR)

```python
import pandas as pd
from sklearn import linear_model

df = pd.read_csv('OrderData.csv')
print(df)

reg = linear_model.LinearRegression()  # Process model
reg.fit(df[['users','orders','age']], df.amount)  # Train model

print("Amount :", reg.predict([[5500,17500,22]])) 
```
