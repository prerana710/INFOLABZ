- Linear regression is a linear approach to modeling the relationship between two variables.

```python
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn import linear_model

df = pd.read_csv('Prices.csv')
print(df)

# y = b0 + (b1*x)
plt.scatter(df.area, df.price, color="r")   
plt.plot(df.area, df.price, color="b")
plt.xlabel('Area')   # Area = Independent = x
plt.ylabel('Price')   # Price = Dependent = y 
plt.show()

reg = linear_model.LinearRegression()  # Process model
reg.fit(df[['area']], df[['price']])  # Train model

print(reg.predict([[1700]]))  # x = 1700 (area)

# y = b0 + (b1*x)
print(reg.coef_)  # b1 = Coefficient
print(reg.intercept_)  # b0 = Intercept
```
