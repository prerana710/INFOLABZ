```python
# From COVID API :  generate horizontal bargraph, state vs totalconfirmed

import requests
from matplotlib import pyplot as plt

url = requests.get("https://data.covid19india.org/data.json")
mydata = url.json()

state = []
confirmed = []

for i in range(0, len(mydata['statewise'])):
    state.append(mydata['statewise'][i]['state'])
    confirmed.append(mydata['statewise'][i]['confirmed'])

plt.figure(figsize=(10, 8))  # Adjust the figure size
bars = plt.barh(state, confirmed, color='skyblue')

plt.title('COVID-19 Confirmed Cases by State in India', fontsize=16)
plt.xlabel('Total Confirmed Cases', fontsize=14)
plt.ylabel('State', fontsize=14)

# Format numbers on the y-axis with commas for better readability
plt.gca().get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.tight_layout()  # Prevents the labels from being cut off
plt.show()
```

```python
# multiple line graph, covid daily data, 565,
# 1 linegraph : newcases (blue), recovered (green), death (red)

import requests
from matplotlib import pyplot as plt

url = requests.get("https://data.covid19india.org/data.json")
data = url.json()

newcase = []
recover = []
death = []

for i in range(0,len(data["cases_time_series"])):
    newcase.append(data["cases_time_series"][i]["dailyconfirmed"])
    recover.append(data["cases_time_series"][i]["dailyrecovered"])
    death.append(data["cases_time_series"][i]["dailydeceased"])
    
plt.plot(newcase, color="blue")
plt.plot(recover, color="green")
plt.plot(death, color="red")

plt.show()
```
